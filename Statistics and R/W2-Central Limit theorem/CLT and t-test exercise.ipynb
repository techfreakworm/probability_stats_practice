## Solution to 1

```r
set.seed(1)
n <- 100
sides <- 6
p <- 1/sides
zs <- replicate(1000, {
    x <-sample(1:sides, n, replace=TRUE)
    (mean(x==6) - p) / sqrt(p*(1-p)/n)
})
qqnorm(zs)
abline(0,1) # confirm it's well approximated with normal distribution
mean(abs(zs) > 2)
```

```r
hist(zs)
```
