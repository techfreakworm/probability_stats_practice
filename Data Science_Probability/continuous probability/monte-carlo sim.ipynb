NOTE: R provides a function to generate normally distributed outcomes <br>
rnorm(n, avg, s) generates n random numbers from the normal distribution with average avg and standard deviation s.

```r
library(tidyverse)
library(dslabs)
data(heights)
x <- heights %>% filter(sex=="Male") %>% pull(height)
```

```r
n <- length(x)
avg <- mean(x)
s <- sd(x)
simulated_heights <- rnorm(n, avg, s)
```

```r
plot(simulated_heights)
```

```r
ds_theme_set()
data.frame(simulated_heights=simulated_heights) %>%
ggplot(aes(simulated_heights)) + geom_histogram(color="black", binwidth=2)
```
