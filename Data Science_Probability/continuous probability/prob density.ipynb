## Probability density comes to save

$$F(a) = \mbox{Pr}(X \leq a) = \int_{-\infty}^a f(x) dx$$

For example, to use the normal approximation to estimate the probability of someone being taller than 76 inches, we can use the probability density

```r
library(tidyverse)
library(dslabs)
data(heights)
x <- heights %>% filter(sex=="Male") %>% pull(height)
```

```r
avg <- mean(x)
s <- sd(x)
1 - pnorm(76, avg, s)
```

NOTE: We can get the probability density function for the normal distribution using the function `dnorm`

```r
plot(prop.table(table(dnorm(x, avg, s))))
```
