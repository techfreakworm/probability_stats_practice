```r
B <- 10^seq(1, 5, len=100)
```

```r
B
```

```r
compute_prob <- function(B, n=22) {
    same_day <- replicate(B, {
        bdays <- sample(1:365, n, replace=TRUE)
        any(duplicated(bdays))
    })
    mean(same_day)
}
```

```r
prob <- sapply(B, compute_prob)
```

```r
plot(log10(B), prob, type='l') # Draw line chart of log of B (number of montecarlo sims)
# Log is generally used to propagate the differences in the middle of the curve
```
